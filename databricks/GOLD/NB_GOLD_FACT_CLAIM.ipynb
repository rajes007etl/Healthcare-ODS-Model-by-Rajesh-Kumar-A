```python
claim_df = spark.table(ods_claim_table).filter(F.col("IS_CURRENT") == "Y")

fact_claim_df = (
    claim_df
    .withColumn("DATE_ID", F.date_format(F.col("CLAIM_DATE"), "yyyyMMdd"))
    .select(
        "CLAIM_ID",
        "MEMBER_ID",
        "PROVIDER_ID",
        "DATE_ID",
        "CLAIM_AMOUNT",
        "CLAIM_YEAR",
        "CLAIM_MONTH",
        "CLAIM_AGE_DAYS",
        "PROCESSING_TIME_DAYS",
        "PAYMENT_LAG_DAYS",
        "CLAIM_BUCKET",
        "STATUS_FLAG"
    )
)

fact_claim_df.write.format("delta").mode("overwrite").saveAsTable(fact_claim_table)
```
